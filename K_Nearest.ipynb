```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('gizi.csv')
df
```

```python
x = df.iloc[:,-1].values
y = df.iloc[:,5].values
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
x_train = x_train.reshape(-1,1)
y_train = y_train.reshape(-1,1)
x_test = x_test.reshape(-1,1)
y_test = y_test.reshape(-1,1)
```

```python
model = KNeighborsClassifier(n_neighbors=3,weights='distance',metric='euclidean')
model.fit(x_train,y_train)
# n_neighbors itu = k
# kenapa distance? karna menggunakan jarak terdekat
# uniform = memanggil  3 label dari lingkaran dan 3 label segitiga
```

```python
y_pred = model.predict(x_test)
```

```python
print('Accuracy :',metrics.accuracy_score(y_test,y_pred))
```
